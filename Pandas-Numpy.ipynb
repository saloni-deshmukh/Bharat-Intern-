```python
import pandas as pd
```

```python
data = {'Name':['John','Anna','Peter','Linda'],
        'Age':[28,34,29,40],
        'City':['New York','Paris','Berlin','London']}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
#1 Write a command to select all rows where the value in the 'Age' column is greater than 30 using `loc`

df.loc[df['Age']>30]
```

```python
#2 How would you select the 'Name' and 'City' columns for rows where the 'Age' column is less than or equal to 29 in the given DataFrame `df`

df.loc[(df['Age']<=29),['Name','City']]
```

```python
#3 Write a command to select the first two rows using `iloc`

df.iloc[0:2,:]
```

```python
#4 How would you select the 'Name' and 'City' columns for the last two rows

df.iloc[-2:,[0,2]]
```

```python
#5 Explain the difference between `loc` and `iloc` in pandas. Provide an example using the DataFrame `df`

# loc - LABEL-based indexing
#iloc - iINTEGER - based indexing

#example : to access the data of 2nd, 3rd row & 2nd, 3rd column

#using LOC

l1 = df.loc[2:3,'Age':'City']
print(l1)

#using ILOC

l2 = df.iloc[2:4, 1:3]
print(l2)
```

```python
#6 Write a command to set the 'City' column to 'Unknown' for all rows where the 'Age' is greater than 30 using `loc

df.loc[df['Age']>30, 'City'] = 'Unknown'
print(df)
```

```python
#7 Using `iloc`, how would you select the second and third rows and the 'Name' and 'City' columns of the DataFrame `df`

df.iloc[1:3, [0,2]]
```

```python
#8 Write a command to select all columns for rows where 'City' is 'Paris' using `loc`

df.loc[df['City']=='Paris']
```

```python
#9 Using `iloc`, how would you select every other row and all columns from the DataFrame `df` starting from the first row?

df.iloc[0:4:2]
```

```python
#10 Using `loc`, write a command to select all rows where 'Name' starts with the letter 'P' in the DataFrame `df`

df.loc[df['Name'].str.startswith('P')]
```
